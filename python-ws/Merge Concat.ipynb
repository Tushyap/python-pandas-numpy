# Merge

```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})

pd.merge(var1, var2, on ="A")
```

```python
pd.merge(var2, var1, on ="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "C":[21,22,23,24,25]})

pd.merge(var1, var2)
```

```python
pd.merge(var1, var2, how = "outer", indicator = True)
```

```python
pd.merge(var1, var2, how = "inner", indicator = True)
```

```python
pd.merge(var1, var2, how = "left", indicator = True)
```

```python
pd.merge(var1, var2, how = "right", indicator = True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6], "B":[21,22,23,24,25]})

pd.merge(var1, var2, left_index = True, right_index = True, suffixes = ("first", "second"))
```

# Concat

```python
sr1 = pd.Series([1,2,3,4,5])
sr2 = pd.Series([21,22,23,24,25])

pd.concat([sr1, sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "B":[21,22,23,24,25]})
pd.concat([d1,d2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d2 = pd.DataFrame({"A":[1,2,3], "C":[21,22,23]})
pd.concat([d1,d2], axis = 1)
```

```python
pd.concat([d1,d2], axis = 1, join = "inner")
```

```python
pd.concat([d1,d2], axis = 1, keys = ["d1", "d2"])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})
pd.concat([d1,d2], axis = 0, keys = ["d1", "d2"])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4,5]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})
pd.concat([d1,d2])
```

```python
d1 = pd.DataFrame({"D":[1,2,3,4,5]})
d2 = pd.DataFrame({"A":[1,2,3,4,5], "C":[21,22,23,24,25]})
pd.concat([d1,d2])
```
