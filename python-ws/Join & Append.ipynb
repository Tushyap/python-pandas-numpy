# Join


```python
import pandas as pd
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"C":[10,20,30,40,50], "D":[21,22,23,24,25]})
var1.join(var2)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]}, index = ["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[10,20,30], "D":[21,22,23]},index = ["a","b","c"])
var1.join(var2)
```

```python
var2.join(var1)
```

```python
var2.join(var1, how = "right")
```

```python
var2.join(var1, how = "inner")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"C":[10,20,30,40,50], "B":[21,22,23,24,25]})
var2.join(var1, how = "outer", lsuffix ="_L", rsuffix ="_R")
```

# Append


```python
var1 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,12,13,14,15]})
var2 = pd.DataFrame({"C":[10,20,30], "D":[21,22,23]})
var1.append(var2, ignore_index = True)
```
