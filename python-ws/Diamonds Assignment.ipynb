# 63. Write a Pandas program to get sample 75% of the diamonds DataFrame's rows without replacement and store the remaining 25% of the rows in another DataFrame.

```python
results = dfData.sample(frac=0.75, random_state=99)
print(results)
print()
print("Store rest 25% of data in another dataframe: ")
results2 = dfData.loc[~dfData.index.isin(results.index), :]
print()
print(results2)
```
