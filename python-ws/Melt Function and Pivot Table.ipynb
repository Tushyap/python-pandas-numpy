# Melt

```python
import pandas as pd

var = pd.DataFrame({"Days":[1,2,3,4,5,6], "Eng":[10,11,12,13,14,15],"Maths":[16,17,18,19,20,21]})
var
```

```python
pd.melt(var)
```

```python
pd.melt(var , id_vars =["Days"])
```

```python
pd.melt(var , id_vars =["Days"], var_name ="Subject", value_name ="Marks")
```

# pivot()

```python
var1 = pd.DataFrame({"Days":[1,2,3,4,5,6],"St_name":["a","b","c","a","b","c"], "Eng":[10,11,12,13,14,15],"Maths":[16,17,18,19,20,21]})
var1
```

```python
var1.pivot(index = "Days", columns = "St_name")
```

```python
var1.pivot(index = "Days", columns = "St_name", values = "Eng")
```

```python
var2 = pd.DataFrame({"Days":[1,1,1,1,2,2],"St_name":["a","b","c","a","b","c"], "Eng":[10,11,12,13,14,15],"Maths":[16,17,18,19,20,21]})
var2
```

```python
var2.pivot_table(index="St_name",columns = "Days", aggfunc = "sum", margins = "True" )
```
